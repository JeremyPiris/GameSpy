```python
zone_dico = []

for i in range(0,merge_03_fusion.shape[0]):
    a = merge_03_fusion['Zone_titre'][i]
    
    if type(a) is str:
        a = a.strip('[]').split(', ')
        a = [x.strip("''") for x in a]
        a = a[:6]

        b = merge_03_fusion['Zone_contenu'][i]
        b = b.strip('[]').split(', ')
        for x in range(0, len(b)):
            b[x] = b[x].strip("''")
            b[x] = b[x].replace('\\n\\n                                        ', '')
            b[x] = b[x].replace('\\n                                    \\n', '')
            b[x] = b[x].replace('\\n', '')
            b[x] = b[x].replace('  ', '')
        b = b[:6]

        dict_test = {
            'Name' : merge_03_fusion['Name'][i],
            'Distributeur' : None,
            'Développeur' : None, 
            'Éditeur' : None, 
            'Franchise' : None, 
            #'Date de sortie FR' : None, 
            #'Date de sortie US' : None, 
            #'Date de sortie JP' : None,
            #'Evaluation PEGI' : None
        }
    
        dict_zone = {}

        for x,y in zip(a,b):
            if x in dict_test.keys():
                dict_a = {x:y}
                dict_zone = dict(dict_zone, **dict_a)

        dict_test = dict(dict_test, **dict_zone)

        zone_dico.append(dict_test)
    else:
        dict_zone = {
            'Name' : merge_03_fusion['Name'][i],
            'Distributeur' : None,
            'Développeur' : None, 
            'Éditeur' : None, 
            'Franchise' : None,
        }
        dict_test = dict(dict_test, **dict_zone)
        zone_dico.append(dict_test)
    
zone_dico
    
```

```python
rajout_df = pd.DataFrame.from_dict(zone_dico)
rajout_df.head()
```
