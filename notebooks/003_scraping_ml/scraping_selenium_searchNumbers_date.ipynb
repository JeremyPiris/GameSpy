```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep
from tqdm import tqdm
import pandas as pd
import pasteboard
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data_col/FinalDataSetMonths.csv', index_col=0)
df = df.dropna(axis = 0, how = 'all', subset = ['Year'])
df['Year'] = df['Year'].astype(int)
df['Mois'] = df['Mois'].astype(int)
for i in df.index:
    if df['Mois'][i] < 10:
        df['DateSortie'][i] = df['DateSortie'][i].replace('0000-01-00', (str(df['Year'][i])+'-0'+str(df['Mois'][i])+'-'+'01'))
    else:
        df['DateSortie'][i] = df['DateSortie'][i].replace('0000-01-00', (str(df['Year'][i])+'-'+str(df['Mois'][i])+'-'+'01'))
df['DateSortie'] = pd.to_datetime(df['DateSortie'])
df = df[df['Year'] > 2004]
```

```python
for i, row in tqdm(df.iterrows(), total=len(df)):
    name = df.loc[i, 'Name']
    month = df.loc[i, 'DateSortie'].month
    y_min = df.loc[i, 'DateSortie'].year - 1
    y_max = df.loc[i, 'DateSortie'].year
    sleep(1)
    driver = webdriver.Chrome(ChromeDriverManager().install())
    sleep(1)
    driver.get(f'https://www.google.com/search?q={name}&tbs=cdr%3A1%2Ccd_min%3A{month}%2F1%2F{y_min}%2Ccd_max%3A{month}%2F1%2F{y_max}')
    sleep(1)
    driver.find_element_by_id('L2AGLb').click()
    action = ActionChains(driver)
    sleep(5)
    driver.find_element_by_id('hdtb-tls').click()
    sleep(2)
    try:
        element = driver.find_element_by_id('result-stats')
        action.double_click(on_element=action.move_to_element_with_offset(element, 55, 20).perform())
        action.perform()
        action.key_down(Keys.COMMAND).send_keys('C').key_up(Keys.COMMAND).perform()
        pb = pasteboard.Pasteboard()
        df.loc[i, 'NbResults'] = pb.get_contents()
    except:
        df.loc[i, 'NbResults'] = 'no_data'
    driver.close()
```
